```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("directorio-de-ejemplo/covid_data.csv")
df[(df['Country_code']=='AR') & (df['New_cases']>0)].tail()
```

```python
filtered_covid = df[(df['Country_code']=='AR') & (df['New_cases']>0) &  (df['Date_reported']>'2022-04-01')]
plt.figure(figsize=(20,5))
plt.title('Covid-19 New_cases ARG')
plt.bar(filtered_covid['Date_reported'],filtered_covid['New_cases'])
plt.xticks(rotation=90)
plt.show()
```
